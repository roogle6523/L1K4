```python
okt = Okt()
```

```python
def noun(fileName):
    #명사 빈도수 파일 불러오기
    f = open('data/insta/{0}.txt'.format(fileName), 'r',encoding = 'utf-8')
    news = f.read()

    noun = okt.nouns(news)

    for i,v in enumerate(noun):
        if len(v)<2:
            noun.pop(i)
    count = Counter(noun)

    
    #명사 빈도 카운트
    noun_cnt=[]
    noun_list = count.most_common(20)
    for v in noun_list:
        noun_cnt.append(v)
    print(noun_cnt)
    return noun_cnt
```

```python
noun("tags")
noun_tags = noun_cnt
```

```python
noun("content")
noun_content = noun_cnt
```
