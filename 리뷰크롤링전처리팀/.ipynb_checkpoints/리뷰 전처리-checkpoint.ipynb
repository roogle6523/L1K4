### 머그워트

```python
dt_mug = pd.read_csv("naver_date_mug.csv", index_col=0, names=["date"], header=0)
rw_mug = pd.read_csv("naver_review_mug.csv", index_col=0, names=["review"], header=0)
pt_mug = pd.read_csv("naver_point_mug.csv", index_col=0, names=["point"], header=0)
```

```python
rw_pt_mug = rw_mug.merge(pt_mug, left_index=True, right_index=True, how="left")
rw_pt_dt_mug = rw_pt_mug.merge(dt_mug, left_index=True, right_index=True, how="left")
rw_pt_dt_mug
```

```python
cols = ["review", "point", "date"]

rw_pt_dt_mug = rw_pt_dt_mug[cols].replace("\n", " ", regex=True)
rw_pt_dt_mug = rw_pt_dt_mug[cols].replace("평점", "", regex=True)
rw_pt_dt_mug
rw_pt_dt_mug.to_csv("rw_pt_dt_mug.csv", header=True, encoding="utf-8-sig")
```
