```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from pandas import Series
import os  
import re
from pandas.api.types import is_numeric_dtype
from pandas.api.types import is_string_dtype
from datetime import datetime
import locale
locale.setlocale(locale.LC_ALL, 'ko_KR')
os.chdir(r'C:\L1K4\리뷰크롤링전처리팀')
```

```python
def naver_open(ingre, filename):
    naver_review = pd.read_csv("C:\L1K4\Data\네이버쇼핑\\naver_{}_review.csv".format(ingre), 
                               header=0, index_col=0, names=["review"])
    
    naver_point = pd.read_csv("C:\L1K4\Data\네이버쇼핑\\naver_{}_point.csv".format(ingre), 
                              header=0, index_col=0, names=["point"])
    
    naver_date = pd.read_csv("C:\L1K4\Data\네이버쇼핑\\naver_{}_date.csv".format(ingre), 
                             header=0, index_col=0, names=["date"])
    
    naver_rw_pt = naver_review.merge(naver_point, left_index=True, right_index=True, how="left")
    naver_ingre = naver_rw_pt.merge(naver_date, left_index=True, right_index=True, how="left")
    
    
    return naver_ingre.to_csv("{}.csv".format(filename), encoding="utf-8-sig")
```

```python
def naver_time(df):
    df["date"] = df["date"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
    df["date"] = df["date"].astype("str")
    df["date"] = df["date"].apply(lambda _: datetime.strptime(_, "%y%m%d"))
```

```python
def time_change(df):
    df["date"] = df["date"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
    df["date"] = df["date"].astype("str")
    df["date"] = df["date"].apply(lambda _: datetime.strptime(_, "%Y%m%d"))
```

```python
def date_sort(df):
    df.sort_values(["date"], ascending=False, inplace=True)
    df.reset_index(drop=True, inplace=True)
```

```python
def ingre_add(df, ing):
    df["ingre"] = ing
```

```python
def concat_file_save(df1, df2, df3, ingre):
    df = pd.concat([df1, df2, df3], ignore_index=True)
    
    return df.to_csv("{}.csv".format(ingre), encoding="utf-8-sig")
```
