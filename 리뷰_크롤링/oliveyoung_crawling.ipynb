```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd 

#find_elements_by_id(‘id명’)
#find_elements_by_class_name(‘class명’)
#find_elements_by_xpath(‘xpath’)
#find_elements_by_tag_name(‘tag명’)
```

```python

def crawling(url,page,filename):
    driver = webdriver.Chrome("C:/Untitled Folder/chromedriver_win32/chromedriver")
    driver.get(url)

    # 리뷰BOX위치를 xpath값을 통해 찾는다.
    review_box=driver.find_element_by_xpath("""//*[@id="reviewInfo"]/a""")
    # 리뷰BOX이벤트를 발생시키기전 브라우저에 나타날때까지 암묵적으로 최대 10까지 기다리게끔 설정
    driver.implicitly_wait(10)
    review_box.click()
    
    text = []
    point = []
    date = [] 
    for num in range(page):
        for num in range(10):
            list3 = driver.find_elements_by_css_selector("div.txt_inner")
            list4 = driver.find_elements_by_css_selector("span.point")
            list5 = driver.find_elements_by_css_selector("span.date")
            for x, i in enumerate(list3,1):
                text.append(i.text)
            for x, i in enumerate(list4,1):
                point.append(i.text)
            for x, i in enumerate(list5,1):
                date.append(i.text)


            page_num = """//*[@id="gdasContentsArea"]/div/div[7]/a[{}]""".format(num+2)
            print(page_num)
            driver.find_element_by_xpath(page_num).click()
            time.sleep(1)
            
    # 갯수확인및 공백컬럼처리       
    print(len(text))
    print(len(point))
    print(len(date))
    point = list(filter(None,point))
    print(len(point))
    
    #데이터프레임형식으로 데이터를 저장
    pd.DataFrame(text[:]).to_csv("data/crawlingdata/txt_review({}).csv".format(filename)
                                 , mode='w',encoding='utf-8-sig')
    pd.DataFrame(point[:]).to_csv("data/crawlingdata/txt_point({}).csv".format(filename)
                                  , mode='w',encoding='utf-8-sig')
    pd.DataFrame(date[:]).to_csv("data/crawlingdata/txt_date({}).csv".format(filename)
                                 , mode='w',encoding='utf-8-sig')
```
