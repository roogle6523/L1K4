```python
def naver_review(url, rw_class, pt_class, out_int, in_int, page_int):
    driver = webdriver.Chrome('C:/myPyCode/analysis/chromedriver')
    driver.implicitly_wait(10)
    driver.get(url)
    
    text = []
    point = []
    for num in tqdm(range(out_int)):
        for num in range(in_int):
            rw = driver.find_elements_by_class_name(rw_class)
            pt = driver.find_elements_by_class_name(pt_class)
            time.sleep(2)

            for x, i in enumerate(rw,1):
                text.append(i.text)

            for x, i in enumerate(pt,1):
                point.append(i.text)


            time.sleep(2)
            page_num = """//*[@id="REVIEW"]/div/div[3]/div/div[2]/div/div/a[{}]""".format(num+page_int)
            driver.find_element_by_xpath(page_num).send_keys(Keys.ENTER)
            print(page_num)
            time.sleep(2)
    driver.close()
```
